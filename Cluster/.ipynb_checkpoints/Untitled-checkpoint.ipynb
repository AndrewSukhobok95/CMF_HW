### Кластеризация

### K-means

Предварительно загрузим sklearn

```python
from sklearn.cluster import KMeans
```

Построение графика внутрегрупповой ошибки от количества кластеров.

```python
error_list = []
for k_num in range(2,20):
    kmeans = KMeans(n_clusters=k_num, random_state=0)
    kmeans.fit(iris['data'])
    error_list.append((1/k_num)*kmeans.inertia_)
```

```python
plt.plot(range(2,20), error_list, 'bo-')
plt.xticks(range(0,21))
plt.axis([1, 20, 0, 80])
```

По графику можно сказать, что оптимальное количество кластеров здесь - 3 или 4 (мы знаем, что 3). Мы видим, что до точки 4 падение достаточно резкое, потом уже более плавное. Что бы ответить на вопрос 3 или 4 - можно попробовать визиулизировать данные для найденных 4 кластеров и посмотреть, что получится. Из графиков ниже видно, что 3 кластера выглядят лучше (менее скомкано).

```python
def kmeans_iris_plot(n):
    '''
    Visualize Iris df (predicted) depending on number of clusters n
    '''
    kmeans = KMeans(n_clusters=n, random_state=0)
    kmeans.fit(iris['data'])
    frame_iris = pd.DataFrame(iris['data'])
    frame_iris.columns = [n.replace(' ', '_') for n in iris['feature_names']]
    frame_iris['species'] = kmeans.labels_
    print('number of clusters:', n)
    sns.pairplot(frame_iris, hue = 'species', vars=frame_iris.columns[:4])
```

```python
kmeans_iris_plot(4)
```

```python
kmeans_iris_plot(3)
```

Сравним предсказание в три кластера с ответами.

```python
kmeans_3 = KMeans(n_clusters=3, random_state=0)
kmeans_3.fit(iris['data'])
kmeans_3.labels_
```

Поскольку kmeans из пакета возвращает ответ, начинающийся с единицы, а имеющиеся ответы начинаются с нуля, то поменяем местами единицы и нули в предсказании.

```python
predict_k = kmeans_3.labels_.copy()
```

Далее нам еще потребуется менять результат кластеризации так, чтоб он подходил под сравнение с таргетом, поэтому завернем этот процесс в функцию.

```python
def pred_change(pred):
    # Gathering clusters sequentially to c_list, as they comes in prediction
    pred = pred.tolist()
    new_c = pred[0]
    c_list = [new_c]
    for c in pred[1:]:
        if c not in c_list:
            c_list.append(c)

    # Replacing numbers like in target
    for i in range(len(pred)):
        for c in list(enumerate(c_list)):
            if pred[i] == c[1]:
                pred[i] = c[0]
                break
    
    return(pred)
```

```python
predict_k = pred_change(predict_k)
```

Сравнение (Процент неверно кластерезированных наблюдений)

```python
compar = predict_k == iris['target']
```

```python
len(compar[compar==False])/len(compar)
```

### Hierarchical clustering

Для иерархической кластеризации используем пакет scipy, так как там больше готовых реализованных метрик.

```python
from scipy.cluster import hierarchy as hc
```

Напишем функцию, которая будет рисовать дендрограмму и диграммы рассевания данных (в данной версии цвета пока различаются, позже недочет будет исправлен), а так же возвращать процент верно кластеризованных наблюдений.

Предварительно - фунция для одной диограммы рассеевания.

```python
# Function for one scatterplot
def iris_sc(position, col_num):
    '''
    position - tuple with coordinates in the subplot
    col_num - column numbers of Iris data set - tuple, where col_num[0] - x, col_num[1] - y
    '''
    sc1 = plt.subplot2grid((4,3),position)
    sc1.set_xlabel(frame_iris.columns[col_num[0]])
    sc1.set_ylabel(frame_iris.columns[col_num[1]])
    plt.scatter(x = frame_iris.iloc[:, col_num[0]], \
                y = frame_iris.iloc[:, col_num[1]], \
                c=frame_iris['species'], \
                cmap=colors)
```

```python
# Notes for arranging the same colors for dendrogram and scetterplots

#z = hc.linkage(iris['data'], method='average')
#pred = hc.fcluster(z, 3, 'maxclust')
#frame_iris = pd.DataFrame(iris['data'])
#frame_iris.columns = [n.replace(' ', '_')[:-5] for n in iris['feature_names']]
#frame_iris['species'] = pred
#hc.leaders(z, pred)
#tree = hc.to_tree(z)
#colors = ListedColormap(sns.color_palette("bright", 4))
#hex_list = []
#for i in range(len(colors.colors)):
#    hex_list.append(mplt.colors.rgb2hex(colors.colors[i]))
```

```python
def hc_iris_plot(link_m, n_clust = 3):
    '''
    Possible linkage methods: single, complete, average, weighted, centroid, median, ward
    
    Now the colors of dendrogram and scatterplot are different
    '''
    
    # define the linkage method
    z = hc.linkage(iris['data'], method=link_m)
    
    # prediction based on choosen metric (linkage)
    pred = hc.fcluster(z, n_clust, 'maxclust')
    
    # making df with prediction for visualization
    frame_iris = pd.DataFrame(iris['data'])
    frame_iris.columns = [n.replace(' ', '_')[:-5] for n in iris['feature_names']]
    frame_iris['species'] = pred
    
    # defining the number of colors in color_paletteof seaborn
    colors = ListedColormap(sns.color_palette("bright", n_clust))
    
    # Percent of successful predictions
    compar = pred_change(pred) == iris['target']
    success = len(compar[compar==True])/len(compar)
    fail = 1 - success
    
    # Printing the input params
    print('Number of clusters:', n_clust)
    print('Linkage method:', link_m)
    print('The success percentage = ', success)
    print('Misclustarization = ', fail)
    
    # Plotting
    plt.figure(figsize=(12, 10)) 

    # Plot dendrogram
    dend = plt.subplot2grid((4,3),(0, 0), colspan=3, rowspan=2)
    hc.dendrogram(z, color_threshold=z[-(n_clust-1),2], above_threshold_color='black')
    dend.set_xticks([])
    plt.title('Dendrogram')

    # Plot 6 scatterplots
    iris_sc((2, 0), (1,0))    
    iris_sc((2, 1), (2,0))
    iris_sc((2, 2), (3,0))
    iris_sc((3, 0), (1,2))
    iris_sc((3, 1), (2,3))
    iris_sc((3, 2), (3,1))

    plt.tight_layout()
```

Как увидим ниже методы centroid и average дали самые близкие к оригинальному разбиению кластеры. Однако, если бы не знали истинных ответов и кластеризовали в слепую, то выбрали бы скорее weighted или ward, так как там расстояние между кластерами больше. Наш выбор был бы уместен, так как с точки зрения близости к истинным кластерам, эти методы уступают упомянутым выше на относительно неольшую величину.

```python
hc_iris_plot(link_m = 'complete', n_clust = 3)
```

```python
hc_iris_plot(link_m = 'average', n_clust = 3)
```

```python
hc_iris_plot(link_m = 'weighted', n_clust = 3)
```

```python
hc_iris_plot(link_m = 'centroid', n_clust = 3)
```

```python
hc_iris_plot(link_m = 'ward', n_clust = 3)
```

### PCA

```python
from sklearn.decomposition import PCA
```

Будем использовать алгоритм k-means для визуализации с помощью PCA.

```python
pca = PCA(n_components=2)
```

Из наших данных получим две компоненты.

```python
pca_data = pca.fit_transform(iris['data'])
```

Теперь центроиды k-means надо также перевести в двумерное пространство. И сразу создадим переменную centr_coords, в которую положим координаты центроидов в удобном для визуализации виде.

```python
centr = pca.transform(kmeans_3.cluster_centers_)
```

```python
centr_coords = list(zip(centr[0], centr[1], centr[2]))
```

Сделаем дата фрейм для удобства при визуализации.

```python
pca_frame = pd.DataFrame(pca_data)
```

```python
pca_frame['species'] = predict_k
```

```python
colors = ListedColormap(sns.color_palette("Paired", 3), 3)

sns.color_palette("hls", 8)
plt.scatter(x = pca_frame[0], \
            y = pca_frame[1], \
            c = pca_frame['species'], \
            cmap = colors)
plt.scatter(x = centr_coords[0], \
            y = centr_coords[1], \
            c = 'r', \
            marker = 'x', s = 200, linewidths=5)
```
