### Data transformation

#### income

```python
def income_tras(df):
    # detemine the patterns
    pattern_av = re.compile('\$\d+K-\$\d+K')
    pattern_high = re.compile('\$\d+Kplus')
    pattern_under = re.compile('\$\d+Kminus')
    
    # transform to numbers
    inc_transformed = []
    for i in range(len(df)):
        if pattern_av.match(df.iloc[i, 6]):
            s = df.iloc[i, 6]
            s = s.replace('$', '').replace('K', '')
            s = s.split('-')
            s = list(map(int, s))
            inc_transformed.append((s[1] - s[0])/2 + s[0])
        elif pattern_high.match(df.iloc[i, 6]):
            s = df.iloc[i, 6]
            s = s.replace('$', '').replace('Kplus', '')
            s = int(s)
            inc_transformed.append(s)
        elif pattern_under.match(df.iloc[i, 6]):
            s = df.iloc[i, 6]
            s = s.replace('$', '').replace('Kminus', '')
            s = int(s)
            inc_transformed.append(s)
        
    # make the df
    df['transformed_income'] = inc_transformed
```

```python
income_tras(df)
```

```python
income_tras(df_test)
```

#### ClinLR and DoleLR

```python
set(df['ClinLR'])
```

```python
pol_scale = list(zip(list(range(-3,4)), ['extLib', 'Lib', 'sliLib', 'Mod', 'sliCon', 'Con', 'extCon']))
```

```python
pol_scale
```

```python
def scale_maker(col, scale):
    scale_list = []
    for c in range(len(col)):
        for s in scale:
            if col[c] == s[1]:
                scale_list.append(s[0])
    return(scale_list)
```

```python
df['ClinLR_scale'] = scale_maker(df['ClinLR'], pol_scale)
```

```python
df['DoleLR_scale'] = scale_maker(df['DoleLR'], pol_scale)
```

```python
df_test['ClinLR_scale'] = scale_maker(df_test['ClinLR'], pol_scale)
```

```python
df_test['DoleLR_scale'] = scale_maker(df_test['DoleLR'], pol_scale)
```

### Classification

```python
from sklearn import svm
from sklearn import cross_validation, metrics
```

```python
train_data, test_data, train_labels, test_labels = \
cross_validation.train_test_split(df_tr.iloc[:, :-1], df_tr.iloc[:, -1], test_size = 0.1)
```

```python
clf = svm.SVC()
clf.fit(train_data, train_labels)
```

```python
svm_labels = clf.predict(test_data)
```

```python
metrics.accuracy_score(test_labels, svm_labels)
```
